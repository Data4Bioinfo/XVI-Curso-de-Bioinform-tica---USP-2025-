### **Exemplo 4:** limpando/trimmando sequ√™ncias üß¨
Considere que temos sequ√™ncias de DNA provenientes de sequenciamento, mas que ainda cont√™m adaptadores... e precisamos remov√™-los!

image.png

````python
# Sequ√™ncias de DNA
DNAsequences = ["GATCGGAAGAGCACACGTCTGAACTCCAGTCACGATAGCTGATCGCTAGGCTAGACTAGTCGTACGTACGTACG", \
                "GATCGGAAGAGCACACGTCTGAACTCCAGTCACGATCGTAGTCGCTCGATAGATATATATATAGTGATGATCGT", \
                "GATCGGAAGAGCACACGTCTGAACTCCAGTCACAGTAGCTGCTCGCTGATCGTAGCGCTGTAGTCGTGTCGTCG"]
````

**Dica:** Para termos est√©ticos, como no caso acima, voc√™ pode continuar o c√≥digo na linha seguinte acrescentando uma barra (\\) ao final da sua linha, mas cuidado, esse deve ser o √∫ltimo caractere da linha, sem espa√ßos ap√≥s!

Primeiramente, vamos ver como fazemos para printar cada sequ√™ncia em uma linha separada?

````python
# Printando separadamente da maneira que sabemos
print(DNAsequences[0])
print(DNAsequences[1])
print(DNAsequences[2])
````

Hmmm.. mas ser√° que √© necess√°rio ficar repetindo toda vez que eu quiser imprimir um elemento da lista?? ü§î

#### **Loop** `for`

Um *loop* permite a execu√ß√£o de uma determinada instru√ß√£o repetidamente, geralmente variando algum elemento

````python
# Para cada linha no arquivo
for line in DNAsequences:
  # Imprima  na tela a linha
  print(line)
````

üí° PythonTutor: https://pythontutor.com/

**Preste aten√ß√£o na indenta√ß√£o e no : anterior**

A indenta√ß√£o √© um *tab* ou 4 espa√ßos simples, mas seja consistente ou ter√° problemas!

image.png

Considerando essa lista de sequ√™ncias, como far√≠amos caso quis√©ssemos fazer um texto imprimindo cada uma delas em uma linha separada, bem como acrescentar a cada linha a frase "√© uma sequ√™ncia de DNA"?

````python
# Para cada linha no arquivo
for line in DNAsequences:
  # Imprima  na tela a linha seguida da frase
  print(line, "√© uma sequ√™ncia de DNA")
````

Agora vamos voltar ao exemplo inicial: sabendo que o adaptador consiste nos 33 primeiros nucleot√≠deos da sequ√™ncia, como faremos para remov√™-lo de cada uma das sequ√™ncias da lista?

````python
# Descobrindo a sequ√™ncia do adaptador para a primeira sequ√™ncia
seq1 = "GATCGGAAGAGCACACGTCTGAACTCCAGTCACGATAGCTGATCGCTAGGCTAGTC"
adapter5end = seq1[0:33]
print(adapter5end)
````

````python
# Para remover o adaptador apenas da primeira sequ√™ncia, far√≠amos uma substitui√ß√£o do adaptador por nada
seq1trimmed = DNAsequences[0].replace(adapter5end,"")

# Imprimindo na tela o antes e o depois
print("Antes da trimmagem:", DNAsequences[0])
print("Depois da trimmagem:", seq1trimmed)
````

E para todas as sequ√™ncias?

````python
# Para cada √≠ndice da lista (0 a 2)
for i in range(3): #o range √© exclusivo ao final
  # Substituindo o adaptador de cada sequ√™ncia
  print(DNAsequences[i].replace(adapter5end,""))
````

### ‚ñ∂ **if e else**
Considere para este exemplo que um gene √© altamente expresso quando a quantifica√ß√£o da express√£o √© maior que 100

````python
geneExpression = 125

# Se a express√£o for maior que 100
if geneExpression > 100: # operador l√≥gico
  # Caso for verdade (True) imprima que o gene √© altamente expresso
  print("O gene √© altamente expresso")
````

### **Exemplo 6**: replicando DNA üß¨

Qual seria a sequ√™ncia de DNA complementar √† seguinte sequ√™ncia de DNA?

````python
sequence = "ATGTTCGGT"
````

### {Dicion√°rios} üìñ
S√£o um outro tipo de estrutura de dados em Python, caracterizados por armazenarem valores pareados

**OBS.:** Est√£o sempre entre chaves { }

image.png

````python
# Criando um dicion√°rio de acordo coma complementaridade de bases
myDict = {"A":"T", "T":"A", "C":"G", "G":"C"}

# Obtendo o valor correspondente a uma chave
print(myDict["A"])
````

````python
# Fazendo o DNA complementar
cDNA = ""

# Percorrendo a sequ√™ncia de DNA
for base in sequence:
  cDNA += myDict[base]

print(cDNA)
````
